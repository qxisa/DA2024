```python
import os
import pandas as pd
import datetime 

city_arr=['Dijon','Paris']
app_arr=['Clash_of_Clans','Fortnite']

all_dfs = []
for city_str in city_arr:
    for app_str in app_arr:
        counter = app_str.count('_')
        for filename1 in os.listdir(f'./data/{city_str}/{app_str}'):
            for filename in os.listdir(f'./data/{city_str}/{app_str}/{filename1}'):
                if f"{city_str}_{app_str}_" in filename and filename.endswith("_DL.txt"):
                    print(filename)
                    day_str = os.path.basename(filename).split('_')[counter+2]
                    day=datetime.datetime.strptime(day_str, '%Y%m%d')
                    times = [day +datetime.timedelta(minutes=15*i) for i in range(96)]
                    times_str = [t.strftime('%H:%M') for t in times]
                    columns = ['tile_id']+times_str
                    df_traffic_dn = pd.read_csv(f'./data/{city_str}/{app_str}/{filename1}/{filename}', sep=' ', names=columns)
                    print(len(df_traffic_dn))
                    df_traffic_dn.to_csv(f'./{city_str}_{app_str}_{day_str}.csv',index=False)
```
